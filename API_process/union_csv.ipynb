```python
import glob
import pandas as pd
```

```python
all_files = glob.glob("..\\data\\process\\*_AFID_60072061.csv")
len(all_files)
```

```python
all_files
```

```python
step = 1
for file in all_files:
    df = pd.read_csv(file,delimiter=',', index_col="id")
    año = int(file.split("\\")[-1][:4])
    df["año"] = año
    print(año, df.shape)

    if step: 
        step = 0
        big_df = df.copy()
    else:
        big_df = pd.concat([big_df, df], ignore_index=True)
```

```python
big_df.shape
```

```python
df_again = pd.read_csv("..\\data\\process\\again.csv",delimiter=',', index_col="id")
```

```python
# Actualizar solo las filas que tienen coincidencia en df_again
big_df = big_df.set_index('url')
df_again_indexed = df_again.set_index('url')

# Solo actualizar las URLs que existen en df_again
common_urls = big_df.index.intersection(df_again_indexed.index)
big_df.loc[common_urls] = df_again_indexed.loc[common_urls]

big_df = big_df.reset_index()
```
