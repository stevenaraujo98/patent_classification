```python
import requests
import pandas as pd
import json
from dotenv import dotenv_values
```

```python
config = dotenv_values(".env")
```

```python
year = 2025
```

```python
url = config["CONTENT_SEARCH_URL"]
start_count = 100
count = 100
apiKey = config["API_KEY"]
api_url = url + "?start=" + str(start_count) + "&count=" + str(count) + "&query=AF-ID%2860072061%29+AND+PUBYEAR+IS+" + str(year) + "&apiKey=" + apiKey
```

```python
response = requests.get(api_url)
data = response.json()
response.status_code
```

```python
data['search-results'].keys()
```

```python
# data['search-results']["opensearch:totalResults"]
# data['search-results']["opensearch:startIndex"]
# data['search-results']["opensearch:itemsPerPage"]
data['search-results']["opensearch:Query"]
```

```python
totalResults = int(data['search-results']["opensearch:totalResults"])
startIndex = int(data['search-results']["opensearch:startIndex"])
itemsPerPage = int(data['search-results']["opensearch:itemsPerPage"])
items_procesados = totalResults - (itemsPerPage - startIndex)
```

```python
totalResults, startIndex, itemsPerPage, items_procesados
```

```python
if items_procesados == totalResults:
    print("Ya no faltan")
else:
    print("Faltan", totalResults-items_procesados)
```

```python
# data['search-results']["link"]
list_results = data['search-results']["entry"]
list_results[0].keys()
```

```python
dict_value = {
    "eid": [],
    "title": [],
    "identifier": [],
    "url": [],
    "urlPage": [],
    "doi": [],
    "aggregationType": [],
    "subtypeDescription": [],
    "affiliation": [],
    "publicationName": [],
    "coverDate": [],
    "sourceId": [],
    "openaccess": [],
    "eIssn/issn/isbn": [],
}
```

```python
for element in list_results:
    if element['openaccess'] == "1":
        print("eid", element['eid'])
        dict_value["eid"].append(element['eid'])
        print("title:", element['dc:title'])
        dict_value["title"].append(element['dc:title'])
        print("identifier:", element['dc:identifier'])
        dict_value["identifier"].append(element['dc:identifier'])
        print("url:", element['prism:url'])
        dict_value["url"].append(element['prism:url'])
        urlPage = "https://www.scopus.com/record/display.uri?eid=" + element['eid'] + "&origin=resultslist"
        print("url page", urlPage)
        dict_value["urlPage"].append(urlPage)
        print("doi:", element['prism:doi'])
        dict_value["doi"].append(element['prism:doi'])

        print("aggregationType:", element['prism:aggregationType'])
        dict_value["aggregationType"].append(element['prism:aggregationType'])
        print("subtypeDescription:", element['subtypeDescription'])
        dict_value["subtypeDescription"].append(element['subtypeDescription'])
        str_affiliation = json.dumps(element["affiliation"])
        print("affiliation:", str_affiliation)
        dict_value["affiliation"].append(str_affiliation)

        print("publicationName:", element['prism:publicationName'])
        dict_value["publicationName"].append(element['prism:publicationName'])
        print("coverDate:", element['prism:coverDate'])
        dict_value["coverDate"].append(element['prism:coverDate'])
        print("source-id:", element['source-id'])
        dict_value["sourceId"].append(element['source-id'])
        print("openaccess:", element['openaccess'])
        dict_value["openaccess"].append(element['openaccess'])

        res_eissn = 0
        try:
            print("eIssn:", element['prism:eIssn'])
            res_eissn = element['prism:eIssn']
        except:
            pass

        try:
            print("issn:", element['prism:issn'])
            res_eissn = element['prism:issn']
        except:
            pass

        try:
            print("isbn:", element['prism:isbn'][0]["$"])
            res_eissn = element['prism:isbn'][0]["$"]
        except:
            pass
        
        dict_value["eIssn/issn/isbn"].append(res_eissn)

        print("-"*120)
```
